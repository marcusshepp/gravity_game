```python


x = linspace(0, 2*pi, 101)
s = sin(x)
c = cos(x)

img = imread('dc_metro.JPG')

close('all')
# 2x2 layout, first plot: sin and cos
subplot(2, 2, 1)
plot(x, s, 'b-', x, c, 'r+')
axis('tight')

# 2nd plot: gridlines, labels
subplot(2, 2, 2)
plot(x, s)
grid()
xlabel('radians')
ylabel('amplitude')
title('sin(x)')
axis('tight')

# 3rd plot, image
subplot(2, 2, 3)
imshow(img, extent=[-10, 10, -10, 10], cmap=cm.winter)

tight_layout()

show()


savefig('my_plots.png')
```
